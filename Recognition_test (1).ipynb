```python
# Instalacija YOLOv8
!pip install -q ultralytics

# Uvoz biblioteka
from ultralytics import YOLO
from google.colab import drive, files
import os

drive.mount('/content/drive', force_remount=True)

model_path = '/content/drive/MyDrive/best.pt'

model = YOLO(model_path)
print("✅ Uspješno učitan trenirani model:", model_path)

# UPLOAD SLIKE
print("\n📤 Uploaduj sliku za detekciju...")
uploaded = files.upload()  # npr. .jpg, .png

#  DETEKCIJA NA SVIM UPLOADOVANIM SLIKAMA
for filename in uploaded:
    print(f"\n🔍 Detekcija na slici: {filename}")
    results = model.predict(source=filename, conf=0.3, verbose=False)
    results[0].show()  # Prikaži rezultat
```
